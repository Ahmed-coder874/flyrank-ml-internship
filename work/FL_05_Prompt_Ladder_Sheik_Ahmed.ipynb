# FL-05: The Prompt Ladder

Name: Sheik Ahmed

## Goal

Improve an AI prompt step-by-step by changing one element at a time and analyzing how the output changes.


## Version 0: Baseline

### Prompt

Build me an AI portfolio website.

### Layer Added

None (baseline)

### Output

The AI created a generic portfolio website with sections like About, Skills, and Projects.

### Analysis

What changed:
- This is the original weak prompt.

What improved:
- No improvement because the prompt was too broad.

What failed:
- The output did not understand my background, target audience, or required features.

What I would try next:
- Add a clearer goal.


# Version 1: Clear Goal

## Prompt

Build me a portfolio website that helps me get AI/ML engineering internships.

## What Changed

Added:
- Clear goal

## Output Improvement

The website focused more on projects and skills instead of being just a personal webpage.

## What Still Failed

It did not know my background or audience.

## Next Improvement

Define the audience.


# Version 2: Defined Audience

## Prompt

Build me a portfolio website for technical recruiters looking for AI/ML interns.

## What Changed

Added:
- Target audience

## Output Improvement

The AI suggested recruiter-focused sections like projects, GitHub links, resume, and achievements.

## What Still Failed

The suggestions were still generic.

## Next Improvement

Add personal context.


# Version 3: Real Context

## Prompt

Build me a portfolio website for technical recruiters looking for AI/ML interns.

I am Sheik Ahmed, an Information Science and Engineering student learning machine learning, artificial intelligence, and software development. I have worked on ML projects, websites, and AI applications.

## What Changed

Added:
- Personal context

## Output Improvement

The AI started suggesting projects and sections matching my actual background.

## What Still Failed

The website structure needed more professional guidance.

## Next Improvement

Specify output format.


# Version 4: Output Format

## Prompt

Create a professional portfolio structure with:

1. Homepage
2. About section
3. Skills section
4. Project case studies
5. Resume section
6. Contact section

For each section explain:
- Purpose
- Content
- Features

## What Changed

Added:
- Required output format

## Output Improvement

The response became organized and easier to implement.

## What Still Failed

The AI did not include enough quality standards.

## Next Improvement

Add quality criteria.


# Version 5: Quality Criteria

## Prompt

Create a professional AI/ML engineering portfolio structure.

Requirements:

- Suitable for internship recruiters
- Modern UI design
- Mobile responsive
- Show real projects
- Explain technical decisions
- Avoid generic buzzwords
- Include measurable outcomes where possible

## What Changed

Added:
- Quality criteria

## Output Improvement

The AI produced a more realistic professional portfolio plan.

## What Still Failed

It still requires my actual project details to make it unique.

## Next Improvement

Add my real project information.


# Final Prompt

You are my AI portfolio mentor.

I am Sheik Ahmed, an Information Science and Engineering student preparing for AI/ML internships.

Help me create a professional portfolio website that proves my ability to build practical AI and software projects.

My portfolio should include:

- Strong homepage introduction
- AI/ML project case studies
- Software engineering projects
- Technical skills
- GitHub integration
- Resume section
- Contact section
- AI chatbot assistant

For every project use:

Problem:
What problem does it solve?

Solution:
What did I build?

Technology:
What tools and frameworks were used?

Result:
What was achieved or learned?

Keep suggestions practical, beginner-friendly, and suitable for recruiters.
