## 1. Ranked actions + reason codes

The ranked action queue prioritizes content pages for human review.

Pages are ranked using observed historical signals such as:
- search impressions
- CTR
- ranking position
- content age
- engagement signals

Reason codes explain why a page appears in the queue:

LOW_CTR:
The page receives impressions but has a lower click-through rate, suggesting the title or snippet may need review.

LOW_RANKING:
The page has weaker ranking position signals and may need content improvement or optimization.

REFRESH_OLD_CONTENT:
The page has not been updated recently and may benefit from a freshness review.

HIGH_IMPRESSIONS:
The page receives search visibility and may have opportunity if optimized.

The ranking is decision support for prioritization, not an automatic action recommendation.


```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
import os


df = pd.read_csv("data/raw/content_refresh_anonymized.csv")


df["ctr"] = df["ctr"].fillna(0)
df["avg_position"] = df["avg_position"].fillna(100)
df["content_age_days"] = df["content_age_days"].fillna(0)
df["impressions_90d"] = df["impressions_90d"].fillna(0)


df["action_score"] = (
    (100 - df["ctr"])
    + df["avg_position"]
    + (df["content_age_days"] / 30)
    + (df["impressions_90d"] / 100)
)


def assign_reason(row):
    if row["ctr"] < 1:
        return "LOW_CTR"
    elif row["avg_position"] > 20:
        return "LOW_RANKING"
    elif row["content_age_days"] > 365:
        return "REFRESH_OLD_CONTENT"
    else:
        return "HIGH_IMPRESSIONS"


df["reason_code"] = df.apply(assign_reason, axis=1)

df["recommended_action"] = "Human Content Review"


queue = df.sort_values(
    "action_score",
    ascending=False
)


queue[
    [
        "content_id",
        "action_score",
        "reason_code",
        "recommended_action"
    ]
].head(10)
```

## 2. Intended use and limits

Intended users:

Content teams and SEO analysts can use this queue to prioritize pages for manual investigation.

The system helps answer:
"Which pages should we review first?"

The system does not answer:
"Which exact changes will improve rankings?"

Limitations:

- Search behavior can change over time.
- Historical patterns may not continue in the future.
- The score does not prove causation.
- Human expertise is required before making content changes.

## 3. Human review + the no-go list

Before taking action, a human reviewer should check:

- Search intent matches the content.
- The page information is still accurate.
- Competitor content changes.
- Technical SEO issues.
- Whether the page serves users well.

Actions that should never be automated:

- Automatically deleting pages.
- Automatically rewriting all content.
- Automatically changing important business information.
- Making decisions without human review.
